```python
def xor_bytes(bytes1, bytes2):
    return bytes([b1 ^ b2 for b1, b2 in zip(bytes1, bytes2)])
```

```python
from string import ascii_lowercase, ascii_uppercase
def brutforce_xor(message):
    l = len(message)
    decrypted_messages = []
    for c in ascii_lowercase+ ascii_uppercase:
        key = (bytes.fromhex((hex(ord(c))[2:])*l))
        try: 
            decrypted_message = (xor_bytes(message, key)).decode("utf-8")
        except: 
            decrypted_message = ""
        decrypted_messages.append(decrypted_message.lower())

    return np.array(decrypted_messages)


```

```python
def get_probs(input_string):
    # Initialize a list of length 52 for lowercase and uppercase letters
    frequencies = np.zeros(26)

    for char in input_string:
        if 'a' <= char <= 'z':
            frequencies[ord(char) - ord('a')] += 1
    epsilon = 0.0001
    probs = frequencies/(frequencies.sum() + epsilon) 
    return probs
def get_score(message):
    return real_prob @ get_probs(message)
v_get_score = np.vectorize(get_score)
v_get_score(np.array(["adsf"]))
```
