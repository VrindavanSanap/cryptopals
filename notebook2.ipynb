```python
from string import ascii_lowercase, ascii_uppercase
def brutforce_xor(message):
    l = len(message)
    decrypted_messages = []
    for c in ascii_lowercase+ ascii_uppercase:
        key = (bytes.fromhex((hex(ord(c))[2:])*l))
        decrypted_message = (xor_bytes(message, key)).decode("utf-8", "ignore")
        decrypted_messages.append(decrypted_message.lower())
    return np.array(decrypted_messages)
```

```python
def get_probs(input_string):
    # Initialize a list of length 52 for lowercase and uppercase letters
    frequencies = np.zeros(26)
    for char in input_string:
        if 'a' <= char <= 'z':
            frequencies[ord(char) - ord('a')] += 1
    epsilon = 0.0001
    probs = frequencies/(frequencies.sum() + epsilon) 
    return probs
def space_count(string):
    score = 0 
    for c in string:
        if c== ' ':
            score += 1
    return score 
def alpha_count(string):
    score = 0 
    for c in string:
        if 'a' <= c<= 'z':
            score += 1
    return score 
epsilon = 0.0001
def alpha_ratio(string):
    return alpha_count(string)/(epsilon + len(string))
def alpha_space_ratio(string):

    return (alpha_count(string) + space_count(string))/(epsilon + len(string))
def get_score(message):
    return alpha_ratio(message) 
v_get_score = np.vectorize(get_score,otypes=[np.float64])
```
